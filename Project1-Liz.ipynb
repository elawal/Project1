```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import requests
import json
url = " https://api.covidtracking.com/v1/states/daily.json"
```

```python
df = pd.DataFrame(requests.get(url).json())
df
```

```python
df_filtered = df[["date","state", "positive","deathIncrease"]]
df_filtered.head()
```

```python
# Bar Chart: Positives by State
state = ["Alaska", "Alabama", "Arkansas", "American Samoa", "Arizona"]
death_increase = [0 , 0, 22, 0, 5]


x = state
y= death_increase

plt.bar(x,y)
plt.xlabel('State')
plt.ylabel("Death Increse")
plt.title('Death Increases by State')
plt.show()
```

```python
# Positive Results by State
state = ["Alaska", "Alabama", "Arkansas", "American Samoa", "Arizona"]
positives = [56686 , 499819, 324818, 0, 826454]


x = state
y= positives

plt.bar(x,y)
plt.xlabel('State')
plt.ylabel("Positive Results")
plt.title('Positive Results by State')
plt.show()
```
